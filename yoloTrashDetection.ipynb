```python
import os
import shutil

# Ścieżka do zipa na Twoim Drive (może być inna, jeśli wrzuciłeś do podfolderu!)
# Zazwyczaj to: /content/drive/MyDrive/dataset.zip
zip_path = '/content/drive/MyDrive/dataset.zip'

# Rozpakowujemy do folderu w Colabie
print("Rozpakowywanie...")
shutil.unpack_archive(zip_path, '/content/dataset')
print("Gotowe!")
```

```python
import yaml

yaml_path = '/content/dataset/data.yaml'

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Nadpisujemy ścieżkę główną
data['path'] = '/content/dataset'
data['train'] = 'train/images'
data['val'] = 'valid/images'
# Jeśli masz test:
if 'test' in data:
    data['test'] = 'test/images'

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("Plik data.yaml zaktualizowany pod Colaba.")
```
